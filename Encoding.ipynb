## one hot encoding with multiple categories

KDD orange with ensemble techniques : finding the most repeated categories and taking the top 10 categories from the list.

```python
data=pd.read_csv("C:/Users/2002m/OneDrive/Documents/CSV FILES/mercedezbenz.csv",usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
for i in data.columns:
    print(len(data[i].unique()))
```

```python
data.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_ten=data.X1.value_counts().sort_values(ascending=False).head(10).index
lst_ten=list(lst_ten)

lst_ten
```

```python
for categories in lst_ten:
    data[categories]=np.where(data['X1']==categories,1,0)
```

```python
lst_ten.append('X1')
```

```python
data[lst_ten].head(10)
```

## mean encoding

```python
data={'SubjectName':['s1','s2','s3','s1','s4','s3','s2','s1','s2','s4','s1'],
      'Target':[1,0,1,1,1,0,0,1,1,1,0]}
  
df = pd.DataFrame(data)
  
print(df)
```

```python
df.count()
```

```python
df.groupby(['SubjectName'])['Target'].mean()
```

```python
m = df.groupby(['SubjectName'])['Target'].mean().to_dict()
  
df['SubjectName'] =  df['SubjectName'].map(m)  
print(df)
```
